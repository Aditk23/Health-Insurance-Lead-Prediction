```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import warnings

warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train_copy = train.copy()
test_copy = test.copy()
```

```python
r1 = train[train['Response']==1].Region_Code.value_counts()

r2 = train.Region_Code.value_counts()
```

```python
train_copy.Region_Code  = train_copy.Region_Code.replace(r1/r2)

test_copy.Region_Code  = test_copy.Region_Code.replace(r1/r2)
```

```python
train_copy.Region_Code = train_copy.Region_Code.fillna(0)

test_copy.Region_Code = test_copy.Region_Code.fillna(0)
```

```python
train_copy['Health Indicator'] = train_copy['Health Indicator'].fillna('X10')

test_copy['Health Indicator'] = test_copy['Health Indicator'].fillna('X10')
```

```python
train_copy['Holding_Policy_Duration'] = train_copy['Holding_Policy_Duration'].replace({np.NaN:0})
train_copy['Holding_Policy_Duration'] = train_copy['Holding_Policy_Duration'].replace({'14+':15})
#train_copy['Holding_Policy_Duration'] = train_copy['Holding_Policy_Duration'].astype('float32')

test_copy['Holding_Policy_Duration'] = test_copy['Holding_Policy_Duration'].replace({np.NaN:0})
test_copy['Holding_Policy_Duration'] = test_copy['Holding_Policy_Duration'].replace({'14+':15})
#test_copy['Holding_Policy_Duration'] = test_copy['Holding_Policy_Duration'].astype('float32')
```

```python
train_copy.Holding_Policy_Type = train_copy.Holding_Policy_Type.fillna(5.0)
test_copy.Holding_Policy_Type = test_copy.Holding_Policy_Type.fillna(5.0)

```

```python
train_copy.isnull().sum()
```

```python
X = train_copy.drop(['ID','Lower_Age','Response'],1)
testing = test_copy.drop(['ID','Lower_Age'],1)
```

```python
y = train_copy.Response
```

```python
X = pd.get_dummies(X,columns=['City_Code','Accomodation_Type','Reco_Insurance_Type', 'Health Indicator',
                     'Is_Spouse','Holding_Policy_Duration','Holding_Policy_Type','Reco_Policy_Cat'])
testing = pd.get_dummies(testing,columns=['City_Code','Accomodation_Type','Reco_Insurance_Type', 'Health Indicator',
                    'Is_Spouse', 'Holding_Policy_Duration','Holding_Policy_Type','Reco_Policy_Cat'])
```

```python
X
```

```python
testing
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled= sc.fit_transform(X)
testing_scaled = sc.fit_transform(testing)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3)

import lightgbm as lgb
from sklearn.metrics import roc_auc_score

clf = lgb.LGBMClassifier()
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
roc_auc_score(y_test, pred, average=None)
```

```python
test_pred = clf.predict_proba(testing_scaled)[:,1]
```

```python
sns.distplot(test_pred)
```

```python
sub = pd.read_csv('sample_submission.csv')

sub['ID'] = test.ID
sub['Response'] = test_pred

sub.to_csv('predictions_lgbm_last.csv',index=False)
```
