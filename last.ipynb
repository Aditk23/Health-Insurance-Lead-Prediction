```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import warnings

warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train_copy = train.copy()
test_copy = test.copy()
```

```python
train
```

```python
c1 = train[train['Response']==1].City_Code.value_counts()

c2 = train.City_Code.value_counts()
```

```python
train_copy.City_Code  = train_copy.City_Code.replace(c1/c2)

test_copy.City_Code  = test_copy.City_Code.replace(c1/c2)
```

```python
train_copy
```

```python
r1 = train[train['Response']==1].Region_Code.value_counts()

r2 = train.Region_Code.value_counts()
```

```python
train_copy.Region_Code  = train_copy.Region_Code.replace(r1/r2)

test_copy.Region_Code  = test_copy.Region_Code.replace(r1/r2)
```

```python
train_copy.Region_Code = train_copy.Region_Code.fillna(0)

test_copy.Region_Code = test_copy.Region_Code.fillna(0)
```

```python
train_copy
```

```python
train_copy['Health Indicator'] = train_copy['Health Indicator'].fillna('X10')

test_copy['Health Indicator'] = test_copy['Health Indicator'].fillna('X10')
```

```python
test_copy.isnull().sum()
```

```python
train_copy
```

```python
train_copy['Holding_Policy_Duration'] = train_copy['Holding_Policy_Duration'].replace({np.NaN:0})
train_copy['Holding_Policy_Duration'] = train_copy['Holding_Policy_Duration'].replace({'14+':15})
train_copy['Holding_Policy_Duration'] = train_copy['Holding_Policy_Duration'].astype('float32').astype('int32')

test_copy['Holding_Policy_Duration'] = test_copy['Holding_Policy_Duration'].replace({np.NaN:0})
test_copy['Holding_Policy_Duration'] = test_copy['Holding_Policy_Duration'].replace({'14+':15})
test_copy['Holding_Policy_Duration'] = test_copy['Holding_Policy_Duration'].astype('float32').astype('int32')
```

```python
test_copy
```

```python
train_copy.Holding_Policy_Type = train_copy.Holding_Policy_Type.fillna(5.0)
test_copy.Holding_Policy_Type = test_copy.Holding_Policy_Type.fillna(5.0)
```

```python
p1 = train[train['Response']==1].Reco_Policy_Cat.value_counts()

p2 = train.Reco_Policy_Cat.value_counts()
```

```python
train_copy.Reco_Policy_Cat  = train_copy.Reco_Policy_Cat.replace(p1/p2)

test_copy.Reco_Policy_Cat  = test_copy.Reco_Policy_Cat.replace(p1/p2)
```

```python
X = train_copy.drop(['ID','Lower_Age','Is_Spouse','Response'],1)
testing = test_copy.drop(['ID','Lower_Age','Is_Spouse'],1)
```

```python
y = train_copy.Response
```

```python
X = pd.get_dummies(X,columns=['Accomodation_Type','Reco_Insurance_Type', 'Health Indicator',
                     'Holding_Policy_Type'])
testing = pd.get_dummies(testing,columns=['Accomodation_Type','Reco_Insurance_Type', 'Health Indicator',
                     'Holding_Policy_Type'])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled= sc.fit_transform(X)
testing_scaled = sc.fit_transform(testing)
```

```python
X
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)


from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
random_forest1 = RandomForestRegressor()
                                       
random_forest1.fit(X_train, y_train) 
pred = random_forest1.predict(X_test)
roc_auc_score(y_test, pred, average=None)
```

```python
sns.distplot(pred)
```

```python
test_pred = pred = random_forest1.predict(testing)
```

```python
sns.distplot(test_pred)
```

```python
sub = pd.read_csv('sample_submission.csv')

sub['ID'] = test.ID
sub['Response'] = test_pred

sub.to_csv('predictions_random_forest.csv',index=False)
```
